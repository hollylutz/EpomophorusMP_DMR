```r
#Estimate log2fold changes for skin/oral etc...

gm_mean = function(x, na.rm=TRUE){
  exp(sum(log(x[x > 0]), na.rm=na.rm) / length(x))
}

geoMeans = apply(DESeq2::counts(hcaf.oral.dds), 1, gm_mean)

hcaf.oral.dds = estimateSizeFactors(hcaf.oral.dds, geoMeans = geoMeans)
hcaf.oral.dds = estimateDispersions(hcaf.oral.dds)
abund = getVarianceStabilizedData(hcaf.oral.dds)

hcaf.oral.dds = DESeq(hcaf.oral.dds, test="Wald", fitType="parametric")

res = results(hcaf.oral.dds)
res.tax = cbind(as(res, "data.frame"), as(tax_table(hcaf.oral.ps)[rownames(res), ], "matrix"))

write.csv(res.tax, file="/home/holutz/projects/BatEcto_MS/Songbird/DESeq/Species_Mapfiles/Oral/log2fold/hcaf.oral.log2fold.csv")

###

geoMeans = apply(DESeq2::counts(mafr.oral.dds), 1, gm_mean)

mafr.oral.dds = estimateSizeFactors(mafr.oral.dds, geoMeans = geoMeans)
mafr.oral.dds = estimateDispersions(mafr.oral.dds)
abund = getVarianceStabilizedData(mafr.oral.dds)

mafr.oral.dds = DESeq(mafr.oral.dds, test="Wald", fitType="parametric")

res = results(mafr.oral.dds)
res.tax = cbind(as(res, "data.frame"), as(tax_table(mafr.oral.ps)[rownames(res), ], "matrix"))

write.csv(res.tax, file="/home/holutz/projects/BatEcto_MS/Songbird/DESeq/Species_Mapfiles/Oral/log2fold/mafr.oral.log2fold.csv")

####

geoMeans = apply(DESeq2::counts(mmin.oral.dds), 1, gm_mean)

mmin.oral.dds = estimateSizeFactors(mmin.oral.dds, geoMeans = geoMeans)
mmin.oral.dds = estimateDispersions(mmin.oral.dds)
abund = getVarianceStabilizedData(mmin.oral.dds)

mmin.oral.dds = DESeq(mmin.oral.dds, test="Wald", fitType="parametric")

res = results(mmin.oral.dds)
res.tax = cbind(as(res, "data.frame"), as(tax_table(mmin.oral.ps)[rownames(res), ], "matrix"))

write.csv(res.tax, file="/home/holutz/projects/BatEcto_MS/Songbird/DESeq/Species_Mapfiles/Oral/log2fold/mmin.oral.log2fold.csv")

###

geoMeans = apply(DESeq2::counts(mnat.oral.dds), 1, gm_mean)

mnat.oral.dds = estimateSizeFactors(mnat.oral.dds, geoMeans = geoMeans)
mnat.oral.dds = estimateDispersions(mnat.oral.dds)
abund = getVarianceStabilizedData(mnat.oral.dds)

mnat.oral.dds = DESeq(mnat.oral.dds, test="Wald", fitType="parametric")

res = results(mnat.oral.dds)
res.tax = cbind(as(res, "data.frame"), as(tax_table(mnat.oral.ps)[rownames(res), ], "matrix"))

write.csv(res.tax, file="/home/holutz/projects/BatEcto_MS/Songbird/DESeq/Species_Mapfiles/Oral/log2fold/mnat.oral.log2fold.csv")

###

geoMeans = apply(DESeq2::counts(raeg.oral.dds), 1, gm_mean)

raeg.oral.dds = estimateSizeFactors(raeg.oral.dds, geoMeans = geoMeans)
raeg.oral.dds = estimateDispersions(raeg.oral.dds)
abund = getVarianceStabilizedData(raeg.oral.dds)

raeg.oral.dds = DESeq(raeg.oral.dds, test="Wald", fitType="parametric")

res = results(raeg.oral.dds)
res.tax = cbind(as(res, "data.frame"), as(tax_table(raeg.oral.ps)[rownames(res), ], "matrix"))

write.csv(res.tax, file="/home/holutz/projects/BatEcto_MS/Songbird/DESeq/Species_Mapfiles/Oral/log2fold/raeg.oral.log2fold.csv")

###

geoMeans = apply(DESeq2::counts(rcliv.oral.dds), 1, gm_mean)

rcliv.oral.dds = estimateSizeFactors(rcliv.oral.dds, geoMeans = geoMeans)
rcliv.oral.dds = estimateDispersions(rcliv.oral.dds)
abund = getVarianceStabilizedData(rcliv.oral.dds)

rcliv.oral.dds = DESeq(rcliv.oral.dds, test="Wald", fitType="parametric")

res = results(rcliv.oral.dds)
res.tax = cbind(as(res, "data.frame"), as(tax_table(rcliv.oral.ps)[rownames(res), ], "matrix"))

write.csv(res.tax, file="/home/holutz/projects/BatEcto_MS/Songbird/DESeq/Species_Mapfiles/Oral/log2fold/rcliv.oral.log2fold.csv")

###

geoMeans = apply(DESeq2::counts(reloq.oral.dds), 1, gm_mean)

reloq.oral.dds = estimateSizeFactors(reloq.oral.dds, geoMeans = geoMeans)
reloq.oral.dds = estimateDispersions(reloq.oral.dds)
abund = getVarianceStabilizedData(reloq.oral.dds)

reloq.oral.dds = DESeq(reloq.oral.dds, test="Wald", fitType="parametric")

res = results(reloq.oral.dds)
res.tax = cbind(as(res, "data.frame"), as(tax_table(reloq.oral.ps)[rownames(res), ], "matrix"))

write.csv(res.tax, file="/home/holutz/projects/BatEcto_MS/Songbird/DESeq/Species_Mapfiles/Oral/log2fold/reloq.oral.log2fold.csv")
```
